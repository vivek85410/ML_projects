# Exploratory Data Analysis

```python
# drop the day,month and year column as not much use in model training
df_copy = df.drop(['day','month','year'], axis = 1)
df_copy
```

```python
# find categories in classes column
df_copy['Classes'].value_counts()
```

```python
df_copy.tail()
```

```python
X = df_copy.drop('FWI',axis=1)
y = df['FWI']
```

# Feature scalling

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state=42)
```

```python
corr = X_train.corr()
corr
```

```python
def correlation(dataset,threashold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threashold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(X_train,0.85)
corr_features
```

```python
X_train
```

```python
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
X_train.shape,X_test.shape
```

```python
# feature scaling or standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Linear Regression Model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

lin_reg = LinearRegression()
lin_reg.fit(X_train,y_train)
y_pred = lin_reg.predict(X_test)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test,y_pred)
```

```python
print("Mean absolute error : ",mae)
print("r2 score : ",score)
```

```python
# Lasso Regression Model
from sklearn.linear_model import Lasso

lasso_reg = Lasso()
lasso_reg.fit(X_train,y_train)
lasso_y_pred = lasso_reg.predict(X_test)
mae = mean_absolute_error(y_test,lasso_y_pred)
score = r2_score(y_test,lasso_y_pred)
```

```python
print("Mean absolute error : ",mae)
print("r2 score : ",score)
```

```python
#Ridge Regression Model
from sklearn.linear_model import Ridge

ridge_reg = Ridge()
ridge_reg.fit(X_train,y_train)
ridge_y_pred = ridge_reg.predict(X_test)
mae = mean_absolute_error(y_test,ridge_y_pred)
score = r2_score(y_test,ridge_y_pred)
```

```python
print("Mean absolute error : ",mae)
print("r2 score : ",score)
```

```python
#ElasticNet Regression Model 
from sklearn.linear_model import ElasticNet

elastic_reg = ElasticNet()
elastic_reg.fit(X_train,y_train)
elastic_y_pred = elastic_reg.predict(X_test)
mae = mean_absolute_error(y_test,elastic_y_pred)
score = r2_score(y_test,elastic_y_pred)
```

```python
print("Mean absolute error : ",mae)
print("r2 score : ",score)
```
